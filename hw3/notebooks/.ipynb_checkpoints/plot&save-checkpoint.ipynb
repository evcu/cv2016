```python
a={}
#cifar
a[9267369]='cifar/default'
a[9270523]='cifar/lr=0.5'
a[9268686]='cifar/lr=0.01/135 epochs'
a[9267393]='cifar/lr=0.5/resampled'
a[9267402]='cifar/lr=0.5/resampled/momentum=0.95'
a[9267409]='cifar/lr=0.5/resampled/momentum=0.75'
```

```python
import matplotlib.pyplot as plt
for k in a:
    x = []
    a_train = []
    a_test = []
    with open('logs/cifar/'+str(k)+'.log','r') as f:
        _ = f.next()
        for l in f:
            b=l.strip().split(',')
            x.append(b[0])
            a_train.append(b[1])
            a_test.append(b[3])
    plt.plot(x,a_train,label="Training Error")
    plt.plot(x,a_test,label="Test Error")
    plt.ylim(0, 20)
    plt.ylabel('Error %')
    plt.xlabel('#Epochs')
    plt.title(a[k])
    plt.legend()
    plt.savefig('c_plots/'+str(k)+'.png')
    plt.show()
```
